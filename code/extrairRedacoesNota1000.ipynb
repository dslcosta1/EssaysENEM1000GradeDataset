```python
# importing the csv module
import csv
```

```python
def create_dic(essay, count, source):
  essay_data = list(filter(None, essay.split("\n")))

  author = essay_data[0]
  them_and_year = essay_data[1]
  anyth = essay_data[1].split(" - ")

  if len(anyth) >= 3:
    prompt_tittle = anyth[1]
    essay_year = anyth[0]
    author_info = "-".join(anyth[2:])
  else:
    prompt_tittle = anyth[0]
    essay_year = anyth[1]
    author_info = ""
  print(essay_year)
  essay_text = '\n'.join(essay_data[2:])

  dic_essay = {
      "id": str(count),
      "author": author,
      "id_prompt": prompt_tittle,
      "essay_year": essay_year,
      "grades": [200, 200, 200, 200, 200, 1000],
      "essay_text": essay_text,
      "source": source,
      "author_info": author_info
  }

  return dic_essay
```

```python
def save_csv(essays, title):
  # field names
  fields = ['id', 'author', 'id_prompt', 'essay_year', 'grades', 'essay_text', 'source', 'author_info']

  # name of csv file
  filename = path_to + title + ".csv"

  # writing to csv file
  with open(filename, 'w') as csvfile:
      # creating a csv dict writer object
      writer = csv.DictWriter(csvfile, fieldnames=fields)

      # writing headers (field names)
      writer.writeheader()

      # writing data rows
      writer.writerows(essays)
```

```python
data_to_csv = []
count = 0

for year in range(2014, 2024):
    filepath = path_to + "essays/"+ str(year) + "/r.txt"

    with open(filepath, 'r') as file:
        lines = file.readlines()
        file_content = ''.join(lines)

    essays = file_content.split("&&")
    essays = list(filter(None, essays))


    for essay in essays:
      print(essay)
      dic_essay = create_dic(essay, count)
      data_to_csv.append(dic_essay)
      count += 1

save_csv(data_to_csv)
```

```python
import pathlib
import textwrap
import time

import google.generativeai as genai

from IPython import display
from IPython.display import Markdown

def to_markdown(text):
  text = text.replace('•', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

```python
def prepare_matadata(clean_metadata, year):
  clean_metadata[1] = str(year) + " - " + themePromptForYear[year] + " - "+ clean_metadata[1]
  clean_metadata[2] = clean_metadata[2] + "&&"
  return "\n".join(clean_metadata)
```

```python
def save_metadata(metadata, year, isTheFirst=False):
  filepath = path_to + "essays/"+ str(year) + "/a.txt"

  if isTheFirst:
    with open(filepath, 'w') as file:
        file.write(metadata)
  else:
    with open(filepath, 'a') as file:
        file.write(metadata)
```
